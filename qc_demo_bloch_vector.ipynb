# visualization Bloch sphere

```python
from qiskit.visualization import plot_bloch_multivector
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np
```

## X gate

```python
# 角度 theta と phi を定義
theta = np.pi / 4
phi = np.pi / 2

# Bloch sphere 上の任意の点を表す量子ビットの状態を定義
initial_state = Statevector([np.cos(theta / 2), np.exp(1j * phi) * np.sin(theta / 2)])

# 初期状態の量子ビットを定義
plot_bloch_multivector(initial_state, title="Initial State")
```

```python
# Xゲートを適用する量子回路を準備
qc = QuantumCircuit(1)
qc.x(0)

# Xゲート適用後の状態ベクトルを取得
final_state = initial_state.evolve(qc)

# 初期状態をプロット

# Xゲート適用後の状態をプロット
plot_bloch_multivector(final_state, title="After X Gate")
```

## Z gate

```python
# Xゲートを適用する量子回路を準備
qc = QuantumCircuit(1)
qc.z(0)

# Xゲート適用後の状態ベクトルを取得
final_state = initial_state.evolve(qc)

# Xゲート適用後の状態をプロット
plot_bloch_multivector(final_state, title="After Z Gate")
```

## H gate

```python
# Xゲートを適用する量子回路を準備
qc = QuantumCircuit(1)
qc.h(0)

# Xゲート適用後の状態ベクトルを取得
final_state = initial_state.evolve(qc)

# Xゲート適用後の状態をプロット
plot_bloch_multivector(final_state, title="After H Gate")
```
